```python
import cv2
from matplotlib import pyplot as plot
import numpy as np
```

```python
an_image = cv2.imread("./data/isolate/DSC_1239.jpg", flags=cv2.IMREAD_GRAYSCALE)
an_image = cv2.resize(an_image, None, fx=0.3, fy=0.3)
cv2.imwrite("./data/isolate/DSC_1239_resized.jpg", an_image)
```

```python
an_image = cv2.imread("./data/isolate/DSC_1239_resized.jpg", cv2.IMREAD_GRAYSCALE)

plot.imshow(an_image, 'gray')
```

```python
ret,thresh1 = cv2.threshold(an_image,32,255,cv2.THRESH_BINARY)
plot.imshow(thresh1, 'gray')
```

```python
th2 = cv2.adaptiveThreshold(an_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,63,4)
plot.imshow(th2, 'gray')
```

```python
th3 = cv2.adaptiveThreshold(an_image, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,63,4)
plot.imshow(th3, 'gray')
cv2.imwrite("./data/isolate/DSC_1239_resized_gaussthreshold.jpg", th3)
```
